To begin, let's import this dataset using the commands below:

---

Данна лабораторна робота присвячена *TreesCount! 2015*. Це широкомасштабна громадська наукова ініціатива, учасники якої збирали дані перепису дерев у Нью-Йорку. Ці результати були зібрані в переписі 2015 року ([Link](https://data.cityofnewyork.us/Environment/2015-Street-Tree-Census-Tree-Data/pi5s-9p35)) та були використанні для планування місцевих ініціатив з розвитку міста.

Для початку імпортуйємо цей набір даних за допомогою команд нижче:






```python
import pandas as pd
lab1_df=pd.read_csv('https://data.cityofnewyork.us/api/views/5rq2-4hqu/rows.csv?accessType=DOWNLOAD')
```

Once we have obtained the new dataset, we can proceed to use head() to view only the beginning of the file.

---

Коли вже отримали новий набір даних, далі можемо почати з використанням head(), щоб переглянути лише початок файлу.






```python
lab1_df.head()
```


Next, we will begin creating a new reduced DataFrame that contains only the following columns: the scientific name of each tree ('spc_latin'), the common name ('spc_common'), the status ('status') which can have values "Alive", "Dead", or "Stump", the health ('health'), and the diameter ('tree_dbh' for intact trees and 'stump_diam' for stumps).

To start, we will create a Python list containing the column names of interest. Then, we will create a new DataFrame named 'lab1_df_reduced' that only contains these columns.

---

Далі розпочнемо створення нового зменшеного DataFrame, що містить лише наступні стовпці: з науковою назвою кожного дерева ('spc_latin'), загальною назвою ('spc_common'), стану - 'status' (значенням є «Живе», «Мертве» або «Пень»), здоров'я - 'health' і діаметр ('tree_dbh' для неушкоджених дерев та 'stump_diam' для пнів).

Для початку **створемо список Python, що містить назви стовпців, які вас цікавлять. Після створемо новий DataFrame під назвою «*lab1_df_reduced*», який містить лише ці стовпці.** 

```python
column_names = ['spc_latin', 'spc_common', 'status', 'health', 'tree_dbh', 'stump_diam']
lab1_df_reduced = lab1_df[column_names]
lab1_df_reduced
```

Next, we will check our new DataFrame for the presence of NaN values that may hinder further analysis. To do this, we run the following code:

---

Далі, перевіримо наш новий DataFrame на наявність значень NaN, які можуть заважати подальшому аналізу. Для цього запускаємо код:

```python
nullValue_counts=lab1_df_reduced.isnull().sum()
nullValue_counts
```

Based on this, we can see that there are 31,619 NaN values in the 'health' column and columns related to names. Let's further investigate, trying to determine the cause.

We will select the rows that have NaN values from our reduced DataFrame using the following command.

---

Виходячи з цього, ми бачимо, що в стовпцях '*health*' і стовпцях пов'язаних з іменами є 31 619 значень NaN. Давайте досліджуватимемо далі, намагаючись з’ясувати причину.

Оберемо рядки, які мають NaN із нашого скороченого DataFrame за допомогою наступної команди.

```python
NaNsOnly=lab1_df_reduced.loc[(lab1_df_reduced['spc_latin'].isnull() ) | (lab1_df_reduced['spc_common'].isnull() ) | (lab1_df_reduced['health'].isnull() ) ]
NaNsOnly.head()
```

Carefully reviewing the following rows, we notice that the status of the first five rows changed from "Alive" to "Dead" and "Stump". Based on this, it appears that there were issues with identifying the species and health status of the dead trees and stumps.

It is unlikely that the "status" column changed randomly, as it would imply that New York is predominantly filled with dead trees and stumps.

Let's run another command so that we can gather more information to determine if these problematic rows were the source of our NaN values.

---

Уважно переглядаємо наступні рядки, та звертаємо увагу, що статус перших п’яти рядків змінився з «*Живий*» на *«Мертвий»* і *«Пень»*. Виходячи з цього, виглядає так, наче були проблеми з визначенням виду та стану здоров’я мертвих дерев і пнів.

Малоймовірно, що стовпець «*статус*» змінився випадково, оскільки інакше це означало б, що Нью-Йорк в основному заповнений мертвими деревами та пнями.

Давайте запустимо ще одну команду, щоб ми могли отримати більше інформації, щоб дізнатися, чи проблемні рядки  були джерелом наших NaN.

```python
pd.value_counts(NaNsOnly['status'])
```

Based on the obtained results, it can be assumed that our reduced NaN DataFrame is mostly not random errors, but rather issues with identifying physically inaccessible trees.

Assuming this, we can now decide how we want to handle the NaN values by exploring different questions about the census. Let's calculate the average diameter for each tree, including dead and/or unidentified ones.

During the calculation of the mean value, we should avoid columns where the diameter is not specified, as including them in our calculations would lead to inaccuracies.

By executing the following command, we can see that there are 278 such rows.

---

На основі отриманих результатів, можна припустити, що наші зменшені NaN DataFrame здебільшого не є випадковими помилками, а натомість є проблеми з ідентифікацією фізично недоступних дерев.

Припускаючи це, тепер ми можемо вирішити, як ми хочемо поводитись з значеннями NaN, досліджуючи різні питання про перепис. Давайте отримаємо середній діаметр для кожного дерева, включаючи мертві та/або неідентифіковані.

Під час обчислення середнього значення слід уникати стовпців, для яких не вказано діаметр, оскільки включення їх у наші розрахунки призведе до їх некорректності. 

Виконуючи наступну команду, ми бачимо, що є 278 таких рядків.

```python
lab1_df_reduced.loc[(lab1_df_reduced['tree_dbh'] == 0) & (lab1_df_reduced['stump_diam'] == 0)]
```

Thus, let's filter out these rows and save a new DataFrame named 'lab1_df_reduced_zerosFiltered'. By using the command provided below, we can confirm that this filtered DataFrame is indeed 278 rows shorter than the previous one.

---

Таким чином, **відфільтруємо ці рядки та зберігаємо новий DataFrame під назвою '_lab1_df_reduced_zerosFiltered_'.** Використовуючи наведену нижче команду, підтверджуємо, що цей відфільтрований DataFrame точно на 278 рядків коротший, ніж попередній.

```python
lab1_df_reduced_zerosFiltered = lab1_df_reduced.loc[(lab1_df_reduced['tree_dbh'] != 0) | (lab1_df_reduced['stump_diam'] != 0)]

print(len(lab1_df_reduced) - len(lab1_df_reduced_zerosFiltered))
```

Next, let's perform the following steps:

1) Group the rows of the filtered DataFrame by the columns 'status', 'spc_latin', 'spc_common', and 'health'.

2) Add 'dropna=False' as an argument to groupby().

3) Use mean() to get the average tree diameter for each group.

4) Save the results as a DataFrame named 'lab1_df_reduced_zerosFiltered_grouped'.

Since we have made several modifications to our original dataset and will continue preparing the DataFrame for further analysis, let's also reset its index column.

---

Далі виконаємо наступні дії:

**1) Групуємо рядки відфільтрованого DataFrame за стовпцями '_status_', '_spc_latin_', '_spc_common_' і '_health_'.**

**2) Додаємо 'dropna=False' як аргумент до groupby().**

**3) Використовуємо mean(), щоб отримати середній діаметр дерева для кожної групи.**

**4) Зберігаємо результати як DataFrame під назвою 'lab1_df_reduced_zerosFiltered_grouped'**

Оскільки ми внесли багато змін до нашого оригінального набору даних, і будемо продовжувати, щоб підготувати DataFrame до цього, також скинемо його стовпці індексу.

```python
lab1_df_reduced_zerosFiltered_grouped = lab1_df_reduced_zerosFiltered.groupby(['status', 'spc_latin', 'spc_common', 'health'], dropna=False)[['tree_dbh', 'stump_diam']].mean().reset_index()
lab1_df_reduced_zerosFiltered_grouped
```

The next step is to create a new column named "diameter" that combines the columns "tree_dbh" and "stump_diam". Since all rows have a non-zero value in one column but not in both, we can safely sum the two columns.

Calling the name of the DataFrame again will display the following:

---

Наступним кроком  **створемо новий стовпець під назвою «*diameter*», який поєднує стовпці «tree_dbh» і «stump_diam».** Оскільки всі рядки мають ненульове значення в одному стовпці, але не в обох, ми можемо підсумовати два стовпці.

Повторюючи виклик імені DataFrame виводить наступне:

```python
lab1_df_reduced_zerosFiltered_grouped['diameter'] = lab1_df_reduced_zerosFiltered_grouped[['tree_dbh', 'stump_diam']].fillna(0).sum(axis=1)
lab1_df_reduced_zerosFiltered_grouped
```

After that, we will execute the following command to remove the columns 'tree_dbh' and 'stump_diam':

---

Після цього **виконцємо таку команду, щоб видалити стовпці '*tree_dbh*' і '*stump_diam*':**

```python
lab1_df_reduced_zerosFiltered_grouped.drop(columns=['tree_dbh', 'stump_diam'], inplace=True)
lab1_df_reduced_zerosFiltered_grouped
```

Our final modification to this DataFrame will involve adding a count for each row. To start, we assign a count of 1 to each row in the reduced but ungrouped DataFrame ('lab1_df_reduced') using the code provided below:

---

Наша остання модифікація цього DataFrame полягатиме в додаванні рахунку для кожного рядку. Для початку **призначаємо кожному рядку у скороченому, але не згрупованому DataFrame ('*lab1_df_reduced*') лічильник 1, використовуючи наведений нижче код:**

```python
lab1_df_reduced_counts=lab1_df_reduced.assign(count = 1)
lab1_df_reduced_counts.head()
```

Next, we perform the following steps on 'lab1_df_reduced_counts':

1) Group the rows of the filtered DataFrame by the columns 'status', 'spc_latin', 'spc_common', and 'health'.

2) Add 'dropna=False' as an argument to groupby().

3) Use sum() to add the count for each group.

4) Save this as a DataFrame named 'lab1_df_reduced_counts_grouped'.

---

Далі виконуємо наступні кроки з *'lab1_df_reduced_counts'*:

**1) Групуємо рядки відфільтрованого DataFrame за стовпцями '_status_', '_spc_latin_', '_spc_common_' і '_health_'.**

**2) Додаємо 'dropna=False' як аргумент до groupby().**

**3) Використовуємо sum(), щоб додати кількість для кожної групи.**

**4) Зберігаємо це як DataFrame під назвою 'lab1_df_reduced_counts_grouped'**

```python
lab1_df_reduced_counts_grouped = lab1_df_reduced_counts.groupby(['status', 'spc_latin', 'spc_common', 'health'], dropna=False)
lab1_df_reduced_counts_grouped = lab1_df_reduced_counts_grouped.sum()
lab1_df_reduced_counts_grouped
```

The next step, to simplify the dataset, is to remove unnecessary columns and reset the index of this DataFrame using the following code:

---

Наступним кроком, щоб спростити набір даних, **видаляємо непотрібні стовпці та скидаємо індекс цього DataFrame за допомогою цього коду:** 

```python
lab1_df_reduced_counts_grouped.drop(columns=['tree_dbh','stump_diam'], inplace=True)
lab1_df_reduced_counts_grouped=lab1_df_reduced_counts_grouped.reset_index()
lab1_df_reduced_counts_grouped.head()
```

As the final modification of our data, we create the final DataFrame named 'lab1_df_final' by merging the "left" (how='left') two DataFrames, 'lab1_df_reduced_counts_grouped' and 'lab1_df_reduced_zerosFiltered_grouped', in that order from left to right.*

After calling its name, we obtain the final DataFrame:

---

В якості останньої модифікації наших даних **створюємо остаточний Dataframe з назвою '*lab1_df_final*', об'єднавши "ліворуч"* (_how='left'_) два DataFrame-и '*lab1_df_reduced_counts_grouped*' і '*lab1_df_reduced_zerosFiltered_grouped*', у цьому порядку зліва направо.**

Після виклику його назви кінцевий DataFrame отримуємо:

```python
lab1_df_final = lab1_df_reduced_counts_grouped.merge(lab1_df_reduced_zerosFiltered_grouped, how='left')
lab1_df_final
```


To improve its appearance in the notebook, we execute the following:

---

Також покращуємо його вигляд у блокноті, виконавши наступне:

```python
lab1_df_final=lab1_df_final.set_index(['status','spc_latin','spc_common','health'])
lab1_df_final
```
